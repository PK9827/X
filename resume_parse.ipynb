````python
import os
from PyPDF2 import PdfReader
from groq import Groq


def extract_resume_test(file_path):
    # Extract text from the resume PDF
    with open(file_path, 'rb') as resume_file:
        pdf = PdfReader(resume_file)
        text = ""
        for page in pdf.pages:
            text += page.extract_text()
        return text
````

````python
import re

def extract_between_backticks(answer):
    # Use regular expression to find the text between two occurrences of three backticks
    match = re.search(r'```(.*?)```', answer, re.DOTALL)
    if match:
        return match.group(1).strip()  # Remove leading/trailing whitespace if any
    else:
        return None

# Example usage
#input_string = "Here is some text ``` split this part ``` and then this part."
extracted_text = extract_between_backticks(answer)
if extracted_text:
    print(f"{extracted_text}")
else:
    print("No text found between backticks.")
````
